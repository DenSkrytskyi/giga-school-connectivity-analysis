# School Connectivity and the GIGA Framework
### Context, Scope and Analytical Approach

This notebook introduces the policy and analytical context of the project.
It outlines the GIGA initiative, defines school connectivity beyond basic Internet access, and explains the rationale behind the selected ITU ICT indicators used in the analysis.

## What is GIGA?

GIGA is a global initiative led by ITU and UNICEF with the objective of connecting every school to the Internet and enabling meaningful digital learning.

Unlike traditional ICT penetration metrics, GIGA focuses on:
- the **quality** of connectivity,
- the **affordability** of access,
- the **reliability** of infrastructure,
- and the **policy and regulatory environment** enabling sustainable school connectivity.

As a result, assessing school connectivity requires a multidimensional analytical approach that goes beyond a single indicator.

## Why school connectivity is not the same as household connectivity

While household and individual Internet access indicators are useful proxies, they do not fully capture the requirements of schools.

Schools typically require:
- stable **fixed broadband** connections,
- sufficient **download throughput and low latency**,
- reliable **backbone and international connectivity**,
- and **affordable pricing** relative to national income levels.

This project therefore combines multiple ICT domains to approximate school connectivity readiness using publicly available ITU data.


## Analytical scope

The analysis focuses on a selected group of countries and a limited set of ITU ICT indicators that are most relevant to school connectivity.

The project is structured around the following domains:
- **Access**: availability of Internet and broadband services,
- **Quality of Service (QoS)**: performance of fixed broadband connections,
- **Backbone / Core infrastructure**: international bandwidth capacity and usage,
- **Affordability**: cost of broadband services relative to income,
- **Governance (contextual)**: universal access policies and financing mechanisms.

This scoped approach ensures analytical depth while remaining aligned with the operational needs of the GIGA initiative.


## Selected ITU ICT indicators

The following indicators are used in the analysis:

**Access**
- Households with Internet access at home
- Active mobile-broadband subscriptions
- Fixed-broadband subscriptions
- Population coverage, by mobile network technology

**Quality of Service (QoS)**
- Average Download Throughput for Fixed Broadband
- Packet Latency for Fixed Broadband
- Service activation time for fixed-broadband service

**Backbone / Core infrastructure**
- International bandwidth usage
- Lit / equipped international bandwidth capacity

**Affordability**
- Fixed-broadband Internet 5GB (% of GNI per capita)
- Data-only mobile broadband basket 5GB (% of GNI per capita)

**Governance (contextual)**
- Universal Access / Service (UAS) policies
- Universal Service Financing mechanisms

## Methodological notes

- The analysis relies exclusively on official ITU data from the ITU DataHub.
- Indicators are harmonized across countries and years where possible.
- No school-level data is used; instead, national-level indicators are interpreted as proxies for school connectivity readiness.
- The project emphasizes interpretability and policy relevance over complex statistical modeling.
